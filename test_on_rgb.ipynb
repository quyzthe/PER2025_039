```python
from utils.saliency import spectral_residual_saliency, compute_bms_saliency, load_rgb_image
import matplotlib.pyplot as plt

img = load_rgb_image('images/apple.jpg')

# Spectral Residual
sr_map = spectral_residual_saliency(img)

# BMS
bms_map = compute_bms_saliency(img)

plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(sr_map, cmap='gray')
plt.title("Spectral Residual")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(bms_map, cmap='gray')
plt.title("BMS Saliency")
plt.axis('off')

plt.show()
```
